```python
import pandas as pd
```

```python
allvisits = pd.read_csv('/Users/aoluwolerotimi/Datasets/train_dejsonified.csv')
```

```python
allvisits.head()
```

```python
allvisits['date'] = pd.to_datetime(allvisits['date'], format="%Y%m%d")
```

```python
# copy record with the earliest date for fullVisitorId into a new dataframe
firstvisits = allvisits.loc[allvisits.groupby('fullVisitorId')['date'].idxmin()]
```

```python
allvisits.shape
```

```python
firstvisits.shape
```

```python
# from first visits, copy these columns into a new dataframe: visitorID, channelGrouping, continent, deviceCategory, pageviews, campaign
firstvisits2 = firstvisits[['fullVisitorId', 'channelGrouping', 'continent', 'deviceCategory', 'pageviews', 'campaign']]
```

```python
# in allvisits, create a column called transaction, which is 1 if transactionRevenue is not null, and 0 if it is null
allvisits['transaction'] = allvisits['transactionRevenue'].notnull().astype(int)
```

```python
allvisits.head()
```

```python
# Calculate the sum of transactions for each fullVisitorID in allvisits
transactions_per_user = allvisits.groupby('fullVisitorId')['transaction'].sum()

# Map this sum to the corresponding fullVisitorID in firstvisits2
firstvisits2['transactions'] = firstvisits2['fullVisitorId'].map(transactions_per_user)
```

```python
firstvisits2.head()
```

```python
# get unique values in transactions column
firstvisits2['transactions'].unique()
```

```python
# transform campaign column so that it contains a 0 if the value is (not set), and 1 otherwise
firstvisits2['campaign'] = (firstvisits2['campaign'] != '(not set)').astype(int)
```

```python
firstvisits2.head()
```

```python
firstvisits2.dtypes
```

```python
# get dummies for channelGrouping, continent, and deviceCategory
firstvisits2 = pd.get_dummies(firstvisits2, columns=['channelGrouping', 'continent', 'deviceCategory'])
```

```python
firstvisits2.head()
```

```python
# check how many records for continent not set
firstvisits2['continent_(not set)'].sum()
```

```python
# check how many records with continent not set and transactions > 0
firstvisits2.loc[(firstvisits2['continent_(not set)'] == 1) & (firstvisits2['transactions'] > 0)]
```

```python
# rename continent_(not set) column to continent_unknown
firstvisits2.rename(columns={'continent_(not set)': 'continent_unknown'}, inplace=True)
```
